```python
#se quiser pode colocar as datas ed inicio e de fim para os dados
#selic = sgs.get({'selic':4032}, start='2010-01-01')

#Do jeito que está aqui traz TODOS OS DADOS
#432 = Taxa de juros - Meta Selic definida pelo Copom  (unidade ao ano, dados diarios)
selic_copom = sgs.get({'SELIC Meta':432})

#Do jeito que está aqui traz TODOS OS DADOS
#11 = Taxa de juros - Selic (unidade ao ano, dados diarios)
selic_diaria = sgs.get({'SELIC Diaria':11})

#Do jeito que está aqui traz TODOS OS DADOS
#1178 = Taxa de Juros - Selic anualizada base 252 (unidade ao ano, dados diarios)
selic_252 = sgs.get({'SELIC Anualiada 252':1178})

#12 = Taxa de Juros - CDI (unidade ao dia, dados diarios)
cdi = sgs.get({'CDI':12})
#cdi['CDI'].format(%)

#se quiser pode colocar as datas ed inicio e de fim para os dados
#inflacao = sgs.get({'ipca':433, 'igp-m':189}, start= '2010-01-01')

#Do jeito que está aqui traz TODOS OS DADOS
#433 = Índice nacionalde preços ao consumidos-amplo (IPCA) (unidade variação mensal, dados mensais)
#189 Índice geral de preços do mercado (IGP-M)(unidade variação mensal, dados mensais)
inflacao = sgs.get({'IPCA':433, 'IGP-M':189})

#O IMA-S representa a evolução, a preço de mercado, de carteira de titulos publicos remunerados pela SELIC (LFT) (unidade indice, dados diários)
ima_s = sgs.get({'IMA-S':12462})

#O IRF-M representa a evolução, a preço de mercado, de carteira de titulos publicos prefixados (LTN e NTN-F) (unidade indice, dados diários)
irf_m = sgs.get({'IRF-M':12461})

#O IRF-M 1 representa a evolução, a preço de mercado, de carteira de titulos publicos prefixados (LTN e NTN-F) com prazo inferiores a um ano (unidade indice, dados diários)
irf_m1 = sgs.get({'IRF-M 1':17626})

#O IRF-M 1+ representa a evolução, a preço de mercado, de carteira de titulos publicos prefixados (LTN e NTN-F) com prazo superiores a um ano (unidade indice, dados diários)
irf_m_mais = sgs.get({'IRF-M +1':17627})

#O IMA-B representa a evolução, a preço de mercado, de carteira de titulos publicos indexados ao IPCA (NTN-B) com prazo superiores a um ano (unidade indice, dados diários)
ima_b = sgs.get({'IMA-B':12466})

#O IMA-B 5 representa a evolução, a preço de mercado, de carteira de titulos publicos indexados ao IPCA (NTN-B) com prazo inferiores a um ano (unidade indice, dados diários)
ima_b5 = sgs.get({'IMA-B 5':12467})

#O IMA-B 5+ representa a evolução, a preço de mercado, de carteira de titulos publicos indexados ao IPCA (NTN-B) com prazo superiores a um ano (unidade indice, dados diários)
ima_b5_mais = sgs.get({'IMA-B 5':12468})

#O IMA-Geral representa a evolução, a preço de mercado, de carteira de titulos publicos federais(unidade indice, dados diários)
ima_geral = sgs.get({'IMA-Geral':12469})

display(selic_copom, selic_diaria, selic_252, cdi, inflacao, ima_s, irf_m, irf_m1, irf_m_mais, ima_b, ima_b5, ima_b5_mais, ima_geral)
```
