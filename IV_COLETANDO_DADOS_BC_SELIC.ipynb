### Codigos para Taxa selic:
#### 11 - Taxa de juros - Selic % a.d
#### 432 - Taxa de juros - Meta Selic % a.a
#### 1178 - Taxa de juros - Selic anualizada base 252 % a.a
#### 4189 - Taxa de juros - Selic acumulada no mês anualizada base 252 % a.a
#### 4390 - Taxa de juros - Selic acumulada no mês % a.m

```python
# TAXA SELIC a.d
#se quiser pode colocar as datas ed inicio e de fim para os dados
selic_d = sgs.get({'Tx Selic % a.d':11}, start='2010-01-01')
selic_d
```

```python
# taxa SELIC META a.a
#432 = Taxa de juros - Meta Selic definida pelo Copom  (unidade ao ano, dados diarios)
selic_mt = sgs.get({'Tx SELIC Meta % a.a':432},start='2010-01-01')
selic_mt
```

```python
# Taxa SELIC Anualizada 252
#Do jeito que está aqui traz TODOS OS DADOS
selic_252 = sgs.get({'Tx SELIC Anualizada 252 % a.a' :1178}, start='2010-01-01')
selic_252
```

```python
# Taxa Selic acumulada no mês anualizada
#Do jeito que está aqui traz TODOS OS DADOS
selic_acum_252 = sgs.get({'Tx SELIC Anualizada 252 % a.a' :4189}, start='2010-01-01')
selic_acum_252
```

```python
# Taxa Selic acumulada no mês
#Do jeito que está aqui traz TODOS OS DADOS
selic_acum_m = sgs.get({'Tx SELIC Anualizada Mes % a.m' :4390}, start='2010-01-01')
selic_acum_m
```

```python
display(selic_d, selic_mt, selic_252, selic_acum_252, selic_acum_m)
```

```python
# juntando os dados de juros em um mesmo dataframe
tx_selic = pd.concat([selic_d, selic_mt, selic_252, selic_acum_252, selic_acum_m], axis=1)

#indicadores_diarios['SELIC Diaria'] = indicadores_diarios['SELIC Diaria'].apply(lambda x: x/100)
#indicadores_diarios['SELIC Meta'] = indicadores_diarios['SELIC Meta'].apply(lambda x: x/100)
#indicadores_diarios['SELIC Anualizada 252'] = indicadores_diarios['SELIC Anualizada 252'].apply(lambda x: x/100)
#indicadores_diarios['SELIC Acum 252'] = indicadores_diarios['SELIC Acum 252'].apply(lambda x: x/100)
#indicadores_diarios['SELIC Acum Mes'] = indicadores_diarios['SELIC Acum Mes'].apply(lambda x: x/100)

tx_selic.fillna('0', inplace= True)

tx_selic
```

```python
tx_selic.to_csv('selic.csv')
```
