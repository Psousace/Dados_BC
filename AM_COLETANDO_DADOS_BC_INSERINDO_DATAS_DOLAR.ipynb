###MOEDAS

```python
from bcb import currency
```

```python
currency.get_currency_list()
moeda = currency.get_currency_list()
moeda.to_excel('Moeda.xlsx')
```

```python
curr = currency.get(['USD', 'EUR', 'CNY'], start= '2010-01-01', end= '2023-01-01')
curr.plot(figsize=(14,6))
```

```python
curr.dropna(inplace=True)
curr
```

```python
curr_normalizada = (curr / curr.iloc[0])*100
curr_normalizada.dropna(inplace= True)
```

```python
curr_normalizada.plot(figsize=(18,8));
```
